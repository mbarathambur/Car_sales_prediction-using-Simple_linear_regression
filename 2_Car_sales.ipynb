### Importing data using pandas

```python
#Import pandas library
import pandas as pd
```

```python
#Load "car_price_prediction_.csv" file using pandas and store in the variable "data"
data=pd.read_csv("car_sales_2.csv")
```

```python
#Print the dataset "data"
data
```

### Seperation of data

```python
#Load Mileage and store in Independent
Independent=data[["Year"]]
```

```python
#Print the variable
Independent
```

```python
#Load Price and store in Dependent
Dependent=data[["Price"]]
```

```python
#Print the variable
Dependent
```

### Train_Test split

```python
#Import sci-kit learn then import function train_test_split
from sklearn.model_selection import train_test_split
```

```python
#Split the Independent using the variable x_train,x_test
#Split the Dependent using the variable y_train,y_test
x_train,x_test,y_train,y_test=train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
#Print the variable
x_train
```

```python
#Print the variable
x_test
```

```python
#Print the varible
y_train
```

```python
#Print the variable
y_test
```

### LinearRegression

```python
#Import LinearRegression Algorithm from sci-kit learn
from sklearn.linear_model import LinearRegression
```

```python
#Store the LinearRegression in variable regression
regression=LinearRegression()
```

```python
#Print the variable 
regression
```

```python
#checking the fit
regression.fit(x_train,y_train)
```

```python
#Store the weight data in the variable "weight"
weight=regression.coef_
```

```python
#Print the varible
weight
```

```python
#Store the bias data in the variable "bias"
bias=regression.intercept_
```

```python
#Print the variable
bias
```

### Evaluation matrix or Validation parameter

```python
#Save the x_test data in the "y_pred" variable 
y_pred=regression.predict(x_test)
```

```python
#Print the variable 
y_pred
```

```python
#Import r2_score from sci-kit learn for testing
from sklearn.metrics import r2_score
```

```python
#Store Accuracy score in the variable "r_score"
r_score=r2_score(y_test,y_pred)
```

```python
#Print the varible 
r_score
```

### Save the Model

```python
#Import pickle library
import pickle
```

```python
#Save the model name
filename="car_sales.sav"
```

```python
#Save the Model into model name
pickle.dump(regression,open(filename,'wb'))
```

```python
#Load the Model
loaded_model=pickle.load(open("car_sales.sav",'rb'))
```

```python
#Predict the result and store in the varible "result"
result=loaded_model.predict([[2024]])
```
